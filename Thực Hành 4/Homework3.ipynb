```python
h = int(input())
for i in range(1, h + 1):
    khoang_trang_dau = " " * (h - i)
    print(khoang_trang_dau, end="")

    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        khoang_trang_giua = " " * (2 * i - 3)
        print("*" + khoang_trang_giua + "*")
```

Bài 14

```python
lon_nhat= None
nho_nhat= None
print("Nhập dãy số và nhập")
while True:
    n=int(input())
    if n==-1:
        break
    if lon_nhat is None:
        lon_nhat=n
        nho_nhat=n
    else:
        if n>lon_nhat:
            lon_nhat=n
        if n<nho_nhat:
            nho_nhat=n
if lon_nhat is not None:
    print("Lớn nhất là:",lon_nhat)
    print("Nhỏ nhất là:",nho_nhat)
else:
    print("Dãy rỗng")
```
