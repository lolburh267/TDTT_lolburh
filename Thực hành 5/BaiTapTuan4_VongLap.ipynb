```python
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if n % i ==0:
            return False
    return True
while True:
    try:
        line = input().split()
        if not line:
            break
        total = 0
        a,b=map(int,line)
        for i in range(a,b+1):
            if is_prime(i):
             total += i
        print(total)
    except E0FError:
        break
```

```python
def daonguoc(n):
    temp = n
    rev = 0
    while temp >0:
        digit = temp % 10
        rev = rev*10+digit
        temp //=10
    return rev
def dadaonguoc(n):
    return n == daonguoc(n)
n = int(input())
dem=0
while not dadaonguoc(n):
    n = n + daonguoc(n)
    dem+=1
print(dem,n)
```

```python
import math
def thanthiet(m):
    cnt=0
    for i in range(1,int(math.sqrt(m))+1):
        if m%i==0:
            cnt+=i+m/i
    cnt-=m
    if(float(math.sqrt(m))-int(math.sqrt(m))==0):
        cnt-=math.sqrt(m)
    return cnt
a,b=map(int,input().split())
if thanthiet(a) == b and thanthiet(b)==a:
    print("true")
else:
    print("false")
```
