# W4A25

```python
n=int(input())
if n==-1:
    print("Nhap so khac")
else:
    lon_nhat=n
    nho_nhat=n
while True:
    n=int(input())
    if n == -1:
        break
    if n> lon_nhat:
        lon_nhat=n
    if n<nho_nhat:
        nho_nhat=n
print(lon_nhat,nho_nhat)
```

# W4A27

```python
s=input()
cnt=0
sotu=False
for char in s:
    if char != " ":
        if not sotu:
            cnt+=1
            sotu=True
    else:
        sotu=False
print(cnt)
```

# W4A29

```python
s=input()
for char in s:
    if 97 <= ord(char)<=122:
        print(chr(ord(char)-32),end="")
    elif 65<=ord(char)<=90:
        print(chr(ord(char)+32),end="")
```

# W4A30

```python
s = input()
s += ","
tong = 0
so_tam = ""
for char in s:
    if char.isdigit(): 
        so_tam += char
    elif char == "," or char == " ":
        if so_tam != "":
            tong += int(so_tam)
            so_tam = ""

print(tong)
```

# W4A31

```python
s=input()
in_hoa=0
in_thuong=0
so=0
for char in s:
    if 65<=ord(char)<=90:
        in_hoa+=1
    elif 97 <= ord(char)<=122:
        in_thuong+=1
    elif 48<=ord(char)<=57:
        so+=1
print(in_hoa,in_thuong,so)
```
