```python
def Max_Of_Two(a,b) -> int:
    if a>b:
        return a
    elif b>a:
        return b
n,m=map(int,input().split())
print(Max_Of_Two(n,m))
```

# W5A4

```python
import math
def shh(a)->int:
    tong=1
    for i in range(2,int(math.sqrt(a))+1):
        if a%i==0:
            tong+=i
            k=a//i
            if k!=i:
                tong+=k
    return tong
n=int(input())
if shh(n)==n:
    print("Là số hoàn hảo")
else:
    print("Không là số hoàn hảo")
```
